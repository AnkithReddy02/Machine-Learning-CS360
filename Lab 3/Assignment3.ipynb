```python
#3)
print('Columnwise mode : ', (df.mode()))
print('Rowwise mode : ', (df.mode(axis = 1)))

# overall mode


print('Overall mode : ')
mydict = {}
for col in df :
  for x in list(df[col]):
    if (x in mydict):
      mydict[x] += 1
    else:
      mydict[x] = 1

l = []

for key,val in mydict.items():
  l.append((val,key))


l.sort(reverse=True)


x,y = l[0]
print(y)
for i in range(1,len(l)):
  x1,y1 = l[i]
  x2,y2 = l[i-1]
  if(x2!=x1):
    break
  
  print(y1)







  

```
