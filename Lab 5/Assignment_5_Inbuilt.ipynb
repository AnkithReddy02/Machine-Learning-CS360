```python

import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston


boston = load_boston()


print(boston.keys())
print(boston.feature_names)


bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names

bos['PRICE'] = boston.target
print(bos)


from sklearn.linear_model import LinearRegression

X = bos.drop(bos.columns[[len(bos.columns)-1]],axis=1)
# print(X)


lr = LinearRegression(copy_X=True, fit_intercept=True, normalize=True)

X_train,X_remaining,Y_train,Y_remaining=train_test_split(X,bos.PRICE,train_size=0.3,random_state=5)
X_validation,X_test,Y_validation,Y_test=train_test_split(X_remaining,Y_remaining,train_size=0.14)

lr.fit(X_train,Y_train)



print(lr)
print(lr.intercept_)
print(lr.coef_)


test_predict = lr.predict(X_test)
validation_predict = lr.predict(X_validation)
train_predict = lr.predict(X_train)


numbers_v=[]
for i in range(len(validation_predict)) : 
  numbers_v.append(i)


plt.scatter(numbers_v,validation_predict,c='b')
plt.scatter(numbers_v,Y_validation,c='g')
plt.xlabel("actual ")
plt.ylabel("predicted ")
plt.title("Graph")
plt.show()

numbers_train = []
for i in range(len(train_predict)) : 
  numbers_train.append(i)

plt.scatter(numbers_train,train_predict,c='b')
plt.scatter(numbers_train,Y_train,c='g')
plt.xlabel("actual ")
plt.ylabel("predicted ")
plt.title("Graph")
plt.show()






print ("Fit a model X_train, and calculate MSE with Y_train:", np.mean((Y_train-lr.predict(X_train)) ** 2))
print ("Fit a model X_train, and calculate MSE with validation:", np.mean((Y_validation - lr.predict(X_validation)) ** 2))
print ("Fit a model X_train, and calculate MSE with X_test, Y_test:", np.mean((Y_test - lr.predict(X_test)) ** 2))




```
